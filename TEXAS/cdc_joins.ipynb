```python
### QUERY 1
```

```python
%%bigquery
select x.SEQNO
from cdc_modeled.OVERALL_Health_Status_2011 as x
INNER JOIN cdc_modeled.OVERALL_Health_Status_2012 as y
ON x.GENHLTH = y.GENHLTH
ORDER BY x.SEQNO
LIMIT 12
```

```python
### The purpose of QUERY 1 is to look at individuals in 2011 and 2012 who had the same overall health
### Overall health is determined by GENHLTH, PHYSHLTH and MENTHLTH
```

```python
### QUERY 2
```

```python
%%bigquery
select x.SEQNO
from cdc_modeled.HEALTH_CARE_2011 as x LEFT OUTER JOIN cdc_modeled.OVERALL_Health_Status_2011 as y
ON x.SEQNO = y.SEQNO
WHERE x._STATE = 48
LIMIT 12
```

```python
### This query uses an outer join. 
### This table is an attempt to find a relationship between health care and overall health status
```

```python
### QUERY 3
```

```python
%%bigquery
select y.SEQNO
from cdc_modeled.DEMOGRAPHIC_2011 as x RIGHT OUTER JOIN cdc_modeled.TOBACCO_USE_2011 as y
ON x._STATE = y._STATE
ORDER BY x.SEQNO
LIMIT 12
```

```python
### This query uses an outer join. 
### This query attempts to find a relationship between demographic and tobacco use (for example most tobacco users are over 18)
```

```python
### QUERY 4
```

```python
%%bigquery
select x.SEQNO
from cdc_modeled.TOBACCO_USE_2011 as x
INNER JOIN cdc_modeled.OVERALL_Health_Status_2011 as y
ON x._STATE = y._STATE
WHERE x._STATE = 48 and y._STATE = 48
LIMIT 12
```

```python
### This query uses an inner join.
### This query looks at data within the same state and brings together information for tobacco use and overall health status.
```

```python
### QUERY 5
```

```python
%%bigquery
select x.SEQNO
from cdc_modeled.OVERALL_Health_Status_2011 as x
INNER JOIN cdc_modeled.OVERALL_Health_Status_2012 as y
ON x.GENHLTH = y.GENHLTH
WHERE x._STATE = 48 and y._STATE = 48
LIMIT 12
```

```python
### This query uses an inner join. 
### This query joins overall health data within two years within the states of Texas and looks for cases where general health is the same.
```

```python
### QUERY 6
```

```python
%%bigquery
select y.EMPLOY
from cdc_modeled.DEMOGRAPHIC_2011 as x
INNER JOIN cdc_modeled.DEMOGRAPHIC_2012 as y
ON x.CHILDREN = y.CHILDREN
WHERE x._STATE = 48 and y._STATE = 48
LIMIT 12
```

```python
### This query is an inner join.
### This query shows the employment status of individuals in Texas with the same number of children in the years 2011 and 2012
```
