```python
# The first query is for the 2011 data table.
# The query below looks at the first 20 entries for when the _STATE is 48. _STATE 48 refers to Texas. 
# The data is ordered by IMONTH in ascending order. IMONTH refers to the month when the participant was interviewed.
# This query separates health data for the state of Texas so we can examine local health issues.  
```

```python
%%bigquery
select * from MASSACHUSETTS_cdc_staging.2011 where _STATE=25 order by IMONTH limit 20
```

```python
# The second query is for the 2012 data table.
# The query shows the first 20 rows of data, where the number of attempts is greater than or equal to 5
# The number of attempts refers to the number of times the participant has completed the surveys in the past
# This dataset includes many individuals who have completed the survey for multiple years in a row. These individuals may be markedly different from the average population and should be examined separately.  
```

```python
%%bigquery
select * from MASSACHUSETTS_cdc_staging.2012 where NATTMPTS>=5 order by IMONTH limit 20
```

```python
# The third query is for the 2013 data table.
# The query shows the Sequence Number (SEQNO) and State for people who were tested positive for HIV (HIVTST6=1), ordered by state. 
# The first 20 rows all are from state 1, which is Alabama
# HIV testing and management is known to be understudied in rural areas, so we separated out the data from Alabama, a predominantly rural state. We can then examine how data from Alabama residents may differ from residents of more urban states. 
```

```python
%%bigquery
select _STATE, SEQNO, HIVTST6 from MASSACHUSETTS_cdc_staging.2013 where HIVTST6=1 AND _STATE=25
```

```python
# The fourth query is for the 2014 data table.
# The query shows the sequence number for the first 20 rows for people who said yes to: "During the past 12 months, have you had either a flu shot or a flu vaccine that was sprayed in your nose?", ordered by the month filed.
# This query allowed us to examine the unique identifiers for respondants who receive the flu vaccine. Accessing a unique identifier is a useful step if we need to request additional data from the CDC. 
```

```python
%%bigquery
select SEQNO, FLUSHOT6 from MASSACHUSETTS_cdc_staging.2014 where FLUSHOT6=1 order by FMONTH limit 20
```

```python
#This query searches the 2015 data. 
#This query shows the sequence data for the first 10 rows where respondants answered "always" to "How often do you use your seatbelt?", ordered by the month filed.
#This data could show patterns between vehicle accidents and seat belt use. 
```

```python
%%bigquery
select * from MASSACHUSETTS_cdc_staging.2015 where SEATBELT = 1 order by FMONTH limit 20
```
